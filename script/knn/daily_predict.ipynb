```python
from ts_knn.knn_ensemble import KnnEnsemble
from ts_knn import get_values
import pandas as pd
from cwms_read.cwms_read import get_cwms
#Ignore the performance warning
import warnings
warnings.filterwarnings('ignore')
```

```python
X_train = pd.read_csv('X.csv', index_col = 0)
y_train = pd.read_csv('y.csv', index_col = 0)
idx = [x for x in X_train.index if x in y_train.index]
X_train = X_train.loc[idx]
y_train = y_train.loc[idx]
```

```python
paths = [
        'BON.%-Saturation-TDG.Inst.1Hour.0.GOES-COMPUTED-REV',
        'BON.Flow-Spill.Ave.1Hour.1Hour.CBT-REV',
        'WRNO.%-Saturation-TDG.Inst.1Hour.0.GOES-COMPUTED-REV'
        ]



raw = get_cwms(paths, lookback = 10, public = True, fill = True)
```

```python
x = raw.dropna()
x.columns = ['forebay', 'spill', 'wrno_tdg']
```

```python
x = x[['spill','forebay']]
```

```python
frames = [X_train, x]
joined = pd.concat(frames, axis = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(joined)
X_train = pd.DataFrame(data = scaler.transform(X_train), index = X_train.index, columns = X_train.columns)
x = pd.DataFrame(data = scaler.transform(x), index = x.index, columns = x.columns)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=1)
#model = KnnEnsemble([3,5,7])
model.fit(X_train, y_train)
```

```python
#X_train[['forebay_t_1','forebay_t_2','forebay_t_3','spill_t_1','spill_t_2','spill_t_3']].head()
```

```python
y_hat = model.predict(x)
y_hat = pd.Series(data =  [x[0] for x in y_hat], index = x.index)
```
